```python
# Allocating independent and dependent features for training and testing
X = final_df.iloc[:,1:]  # All feautres except Selling Price
Y = final_df.iloc[:,0]  # Selling price features
print("Shape of the Indepedent features :- ", X.shape)
print("Shape of the Indepedent features :- ", Y.shape)
```

```python
# Features imporatance
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(X,Y)
```

```python
# Now if you want to see which feautres having more importance
print(model.feature_importances_)
```

```python
# Plot graph to know about the top 5 features
feat_impt = pd.Series(model.feature_importances_, index = X.columns)
feat_impt.nlargest(5).plot(kind='barh')
plt.show()
```

```python
# Let's seprate trainig and testing data
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=0)
```

```python
# Let's call the model Random forest
from sklearn.ensemble import RandomForestRegressor

# creating object of model
regressor = RandomForestRegressor()
```

```python
# Let's create hyper parameter tunning to enhance model accuracy
import numpy as np
n_estimators = [int(x) for x in np.linspace(start = 100, stop=1200, num=12)]
n_estimators
```

```python
# Let's call RandomizedSearchCV to implement hyper parameter tunning
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
# Let's create the random grid
# Use the random grid to search for best hyperparameters
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}

print(random_grid)
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations

from sklearn.model_selection import RandomizedSearchCV
rf_random = RandomizedSearchCV(estimator = regressor, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
# Now let's run the model on train data set
rf_random.fit(X_train, Y_train)
```

```python
rf_random.best_params_
```

```python
rf_random.best_score_
```

```python
predictions = rf_random.predict(X_test)
predictions
```

```python
# Let's create the displot of the prediction to see the distribution
sns.distplot(Y_test-predictions)
```

```python
plt.scatter(Y_test,predictions)
```

```python
# Let's check the accuracy of the model
from sklearn import metrics

print('MAE:', metrics.mean_absolute_error(Y_test, predictions))
print('MSE:', metrics.mean_squared_error(Y_test, predictions))
print('RMSE:', np.sqrt(metrics.mean_squared_error(Y_test, predictions)))
```
